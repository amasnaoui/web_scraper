```python
from bs4 import BeautifulSoup
import requests
import os
import shutil
```

```python
url = "https://www.autodeknudt.com/switch/en/A"
```

```python
def creat_car_dir(path):
  if os.path.isdir(path):
    shutil.rmtree(path)
  os.mkdir(path)
  print(path + ' created ..')
```

```python
def unique(list0):
    
    unique_list = []
  
    for x in list0:
        if x not in unique_list:
            unique_list.append(x)
    return unique_list
```

```python
def get_car(url):
  re = requests.get(url)
  soup =  BeautifulSoup(re.text, 'html.parser')
  links = soup.find_all('img')

  cars = []
  for link in links:
    if "/pictures/BE870635970/" in str(link.get('src')) and str(link.get('src'))[-4:] == '.JPG':
      cars.append(link.get('src'))
  return cars
```

```python
def get_links(url):
  re = requests.get(url)
  soup =  BeautifulSoup(re.text, 'html.parser')
  links = soup.find_all('a', href=True)
  links_list = []
  for link in links:
    if "car-details/car/" in (link.get('href')):
      links_list.append("https://www.autodeknudt.com/"+str(link.get('href')))
  return unique(links_list)
#print(get_links(url))
```

```python
def run(url):
  links = get_links(url)
  cars_names = []
  for link in links:
    car_name = link.split('/')[-1]
    cars_names.append(car_name)
    # print(car_name)
    creat_car_dir("/content/drive/MyDrive/autodeknudt.com/"+car_name+"-v"+str(cars_names.count(car_name)))
    images = get_car(link)
    for image in images:
      print(f"adding {image} ... ")
      with open("/content/drive/MyDrive/autodeknudt.com/"+car_name+"-v"+str(cars_names.count(car_name))+"/"+image.split('/')[-1], 'wb') as f:
       im = requests.get(image) ###âˆš
       f.write(im.content)
  # print(len(cars_names))
```

```python
run(url)
```
